```python
df.columns=['data/hora','rock', 'hip hop', 'mpb', 'reggae', 'funk', 'sertanejo', 'gospel', 'eletronica' , 'professor' , 'faixa etaria']
```

```python
# dropando as colunas(removendo)
df.drop(columns=['data/hora'])
```

```python
filtrado=df.loc[df['professor']=='Luciano']
# O método loc armazenado em outra variável, ele seleciona apenas as linhas onde as colunas for igual a professor.
```

```python
filtrado.drop(columns=['data/hora','faixa etaria']).sum()
# Um drop foi feito para remover as colunas data/hora e faixa etaria do dataset ( perguntar se era para remover a faixa etaria )
```

```python
somapontuacao=filtrado[['rock', 'hip hop', 'mpb', 'reggae', 'funk', 'sertanejo', 'gospel', 'eletronica']].sum()
# A lógica aqui é que eu estou selecionando todas as colunas, por nao ter colunms nem um sinal de =, eu selecionei todas e depois fiz a soma de cada um.
# os dois colchetes, permite selecionar multiplas colunas.
# Apenas um colchete, permite que eu selecione apenas uma coluna: rock por exemplo.
```

```python
somapontuacao.plot(kind='barh')
# Aqui por fim, gera o gráfico.
```
